### step 5 - The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
#calculate statistics around product prices for each loyalty category
df.groupby('loyalty_flag').agg({'prices':['mean']})
```

### step 6 - The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders.

```python
#using transform(), add a new column with the average prices paid for each user
df['average_price'] = df.groupby('user_id')['prices'].transform(np.mean)
```

```python
#using loc(), create a spending flag based on average prices paid for each user

#establish conditions
df.loc[df['average_price'] >= 10, 'spending flag'] = 'High spender'
```

```python
#establish conditions
df.loc[df['average_price'] < 10, 'spending flag'] = 'Low spender'
```

```python
#call relevant columns to check 
df[['user_id', 'average_price', 'spending flag']].head(50)
```

```python
# count values of relevant columns to check
df['spending flag'].value_counts(dropna = False)
```

### step 7 - In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. 

```python
#using transform(), add a new column with the median order frequency of each user 
df['order_frequency'] = df.groupby('user_id')['days_since_prior_order'].transform(np.median)
```

```python
#using loc(), create an order frequency flag based on median days between orders for each user

#establish conditions
df.loc[df['order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
#establish conditions
df.loc[(df['order_frequency'] <= 20) & (df['order_frequency'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
#establish conditions
df.loc[df['order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
#call relevant columns to check
df[['user_id', 'order_frequency', 'order_frequency_flag']].head(50)
```

```python
# count values of relevant columns to check
df['order_frequency_flag'].value_counts(dropna = False)
```
