# Manipulate Data (Exercise 5.8 Walk-through)

```python
df = df_ords_prods_merge[:1000000]
```

```python
#group dataframe by department_id and calculate mean of order number
df.groupby('department_id').agg({'order_number':['mean']})
```

```python
#alternative code to calculate mean of order number 
df.groupby('department_id')['order_number'].mean()
```

```python
#aggregate multiple times
df.groupby('department_id').agg({'order_number' : ['mean', 'max', 'min']})
```

```python
#aggregating data with transform()

df_ords_prods_merge['max order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# deriving new columns with loc()

df_ords_prods_merge.loc[df_ords_prods_merge['max order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max order'] <= 40) & (df_ords_prods_merge['max order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max order'] <= 10, 'loyalty_flag'] = 'New customer'
```
